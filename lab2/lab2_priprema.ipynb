b) Koristeći Newtonov interpolacijski polinom $P_{n}(x)$ koji interpolira sljedećih $n+1$ točaka 
$$\{(x_j,f(x_j)), j=0,1,2,\ldots n\}$$
koristeći uniformno raspoređene čvorove $x_j=-1+\frac{2j}{n}$ aproksimirajte funkciju $f(x)$.

Nadalje, uzmite $n=5,7, 11, 13$ te prikažite na istom grafu funkciju i interpolacijski polinom koristeći vektor s komponentama $z_j=-1+\frac{2j}{5n}, j=0,1,\ldots 5n$, u čijim komponentama prikazujete vrijednosti funkcije i polinoma.
Za svaki $n$ nacrtajte i pogrešku $e(x)=|f(x)- P_n(x)|$

```python
import numpy as np
from numpy import array
import matplotlib.pyplot as plt

m = np.vectorize(lambda x : 1 / (1 + 25 * x * x))

def evalPoly(c,x,xu):
    n = len(x) - 1 # Degree of polynomial
    p = c[n]
    for k in range(1,n+1):
        p = c[n-k] + (xu -x[n-k])*p
    return p

def polycoef(x,y):
    n = len(x) # Number of data points
    c = y.copy()
    for j in range(1,n):
        c[j:n] = (c[j:n] - c[j-1])/(x[j:n] - x[j-1])
    return c

def NewtonInterp(x,y,xu):
    c = polycoef(x,y)
    yu = evalPoly(c,x,xu)
    return yu

f = lambda x : 1 / (1 + 25 * x * x)

def plot(n):
    xl = []
    yl = []
    for j in range(5*n+1):
        x = -1 + (2*j/(5*n))
        y = f(x)
        xl.append(x)
        yl.append(y)
    return xl,yl 

x,y = plot(5)
x,y = array(x), array(y)

xp = np.linspace(-1, 1, 1000)
yu = NewtonInterp(x,y,xp)

u = m(xp)

fig, ax = plt.subplots(figsize=(8, 8))   
plt.plot(x, y, 'bo', label="Podaci (n = 5)")
plt.plot(xp, yu,label="Interpolacijski polinom")
plt.plot(xp, u, 'r', label="Funkcija")
ax.legend(loc='upper right')
plt.show()

fig, ax = plt.subplots(figsize=(8, 8))   
plt.plot(xp,abs(yu-u),'r',label = "Apsolutna pogreška")
ax.legend(loc='upper right')
plt.show()
```

c) Napišite program koji računa polinom $\hat P_m(x) (m= 2, 5,10)$ (možete koristiti ugrađene funkcije) tako da aproksimira 100 random odabranih točaka oblika 
$(x,f(x))$ za $x\in [-1,1]$, u smislu najmanjih kvadrata. Prikažite polinome $\hat P_m(x)$ za različite $m$ i funkciju $f(x)$ na istoj slici.
Za svaki $m$ nacrtajte i pogrešku $e(x)=|f(x)-\hat P_m(x)|$ 

```python
import numpy as np
from numpy import array
import matplotlib.pyplot as plt
from random import random

f = lambda x : 1 / (1 + 25 * x * x)

x = []
y = []

for _ in range(100):
    x.append(random() * 2 - 1)

x.sort()

for i in range(100):
    y.append(f(x[i]))

x = array(x)
y = array(y)

fig, ax = plt.subplots(figsize=(8, 8))   
plt.plot(x, y, 'bo', label="Podaci")

xx = np.linspace(-1, 1, 20000)

m = np.vectorize(f)
u = m(xx)

plt.plot(xx, u, 'r', label="Funkcija")

a = np.polyfit(x, y, 2)
z = np.linspace(-1, 1, 1000)
pp = np.polyval(a,z)
plt.plot(z, pp, label=r'$p_2(x)$')
plt.legend(loc='upper left')
plt.show()

fig, ax = plt.subplots(figsize=(8, 8))   
plt.plot(z, abs(np.polyval(a, z) - m(z)),'r',label = "Apsolutna pogreška")
ax.legend(loc='upper right')
plt.show()
```
